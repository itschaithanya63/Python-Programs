```python
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reversed number:", rev)
```

```python
 num = int(input("Enter a number: "))
count = 0

while num > 0:
    num = num // 10
    count += 1

print("Number of digits:", count)
```

```python
num=int(input("Enter a number"))
def fact(num):
    fact=1
    for i in range(1,num+1):
        fact*=i
    return fact
print("Factorial of",num,"is",fact(num))
```

```python
num=int(input("Enter a number"))
def isprime(num):
    if num<=1:
        return "not Prime"
    for i in range(2,num):
        if num%i==0:
            return "not prime"
    return "prime"
print(num, "is",isprime(num))
```
