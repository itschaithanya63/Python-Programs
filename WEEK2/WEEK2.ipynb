```python
m1=float(input("Enter subject1 marks"))
m2=float(input("Enter subject2 marks"))
m3=float(input("Enter subject3 marks"))
tot=m1+m2+m3
avg=tot/3
if avg>=90:
    print("Grade=A")
elif avg>=80:
    print("Grade=B")
elif avg>=70:
    print("Grade=C")
elif avg>=60:
    print("Garde=D")
elif avg>=50:
    print("Grade=E")
else:
    print("FAIL!")
```
