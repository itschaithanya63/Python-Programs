```python
li= [10, 3.5, "hello", True]
print("List:",li)
print("Type of list:", type(li))
for item in li:
    print(item, "->", type(item))
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
# Concatenation
print("Concatenation:", a + b)
# Repetition
print("Repetition:", a * 3)
# Membership check
print("Is 2 in list a?", 2 in a)
# Length
print("Length of list a:", len(a))
```

```python
nums = [10, 5, 20, 8, 15]
print("Max:", max(nums))
print("Min:", min(nums))
print("Sum:", sum(nums))
print("Ascending:", sorted(nums))
print("Descending:", sorted(nums, reverse=True))
print("Index of 8:", nums.index(8))
```

```python
lst = [10, 20, 30]
lst.append(40)
print("After append:", lst)

lst.insert(1, 15)
print("After insert:", lst)

lst.remove(20)
print("After remove:", lst)

lst.pop()
print("After pop:", lst)

lst.reverse()
print("After reverse:", lst)
```
